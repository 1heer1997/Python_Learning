## Categorical Plots

Seaborn also helps us in doing the analysis on Categorical Data points. In this section we will discuss about

- boxplot
- violinplot
- countplot
- bar plot

### Count Plot 

```python
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
```

```python
df.head() # we have sex, smoker, day and time as categorical features
```

```python
## count plot 

# shows the counts of each observations in each categorical bin using bars

sns.countplot('sex', data=df)
```

```python
df['sex'].value_counts()
```

```python
sns.countplot('smoker', data=df)
```

```python
sns.countplot('day', data=df) # default x axis is taken for categorical feature

# most of the people goes on sat sun to restaurant for eating
```

```python
df['day'].value_counts()
```

```python
# Horizontal observation plot when y axis - categorical feature given

sns.countplot(y = 'sex', data=df)
```

### Bar Plot

```python
# in bar plot we have to give both axis as compared to count plot

sns.barplot(x='total_bill', y='sex',data=df)

# in the case of male - total bill cpme somewhere around 20 and in female - 17
```

```python
sns.barplot(x='total_bill', y='smoker',data=df)

# In the case of smoker- max bill come somwhere around more than 20 and 
# in the case of non-smoker - bill 17 or 18
```

### Box Plot 

A Box and whisker plot (sometimes called a boxplot) is a graph that presents information from a five-number summary.

```python
sns.boxplot('smoker','total_bill', data=df)
```

```python
sns.boxplot('sex','total_bill', data=df)
```

Male has also given bill more than 50 doller which is a kind of an outlier.

```python
sns.boxplot(x="day",y="total_bill", data=df, palette = 'rainbow')
```

```python
sns.boxplot(data=df, orient='v') # considering all the features here
```

```python
# categorize my data based on some other categories

sns.boxenplot(x = "total_bill", y="day", hue="smoker", data=df) # classify points with respect to smoker category
```

## Violin Plot

Violin plot helps us to see both distribution of data in terms of kernel density estimation and the box plot.

```python
sns.violinplot(x="total_bill", y="day", data=df, palette = "rainbow")
```
